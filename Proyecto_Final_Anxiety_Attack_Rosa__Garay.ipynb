**Abstracto con Motivación y Audiencia**

Este dataset contiene información recopilada de diversos y posibles factores que pueden provocar ataques de ansiedad; entre algunos de ellos, se tiene la edad, nivel de estrés, actividad física, consumo de alcohol,entre otros.El análisis que se obtendrá va dirigido para profesionales del rubro de la salud mental, puesto que podrían brindar un diagnóstico mucho más certero y en menor tiempo, asimismo el análisis es viable para cualquier persona que desea conocer si tiene un indicio de padecer un ataque de ansiedad y pueda buscar ayuda profesional. **Descripción de variables**

Diccionario de variables del dataset Anxiety Attack:

1. ID: Identificador único por persona, el cuál permite también conocer el total de participantes en la base de datos.
2. Age (edad): Edad de las personas que formar parte del dataset en años
3.Gender (Género): Género de las personas (masculino, femenino u otro)
4. Occupation (Ocupación): Tipo de empleo o actividad del participante.
5. Sleep Hours (Horas de sueño): Cantidad promedio de horas de sueño de cada persona por noche.
6. Caffeine Consumption (mg/day) (Consumo de cafeína en mg/día): Cantidad de cafeína consumida diariamente.
7. Alcohol Consumption (drinks/week) (Consumo de alcohol en bebidas/semana): Número de bebidas alcohólicas consumidas en una semana.
8. Physical Activity (hours/week) (Actividad física en horas/semana): Horas de actividad física realizadas por semana.
9. Stress Level (1-10) (Nivel de estrés): Nivel de estrés percibido por el participante en una escala de 1 a 10.
10. Family History of Anxiety (Historial familiar de ansiedad): Indica si el participante tiene antecedentes familiares de ansiedad (Sí/No).
11. Therapy (Terapia): Indica si la persona asiste a terapia psicológica (Sí/No).
12. Medication (Medicación): Indica si una persona toma medicación para la ansiedad (Sí/No).
13. Severity of Anxiety Attack (1-10) (Severidad del ataque de ansiedad): Intensidad del ataque de ansiedad experimentado en una escala de 1 a 10.
14. Daily Screen Time (hours/day) (Tiempo de pantalla diario en horas/día): Cantidad de horas que el participante pasa frente a pantallas diariamente.
15. Diet Type (Tipo de dieta): Tipo de alimentación seguida por el participante (Existen los siguiente tipos: vegetariana, balanceada, alta en grasas, etc.).
16. Social Support (1-10) (Apoyo social): Nivel de apoyo social percibido en una escala de 1 a 10.
17. Work Hours (hours/week) (Horas de trabajo semanales): Número de horas que la personas trabaja por semana.
18. Financial Stress (1-10) (Estrés financiero): Nivel de estrés relacionado con la situación económica que presenta el participante en una escala de 1 a 10.
19. Major Life Events (past year) (Eventos importantes en el último año): Número de eventos significativos o importantes que el participante ha experimentado en el último año (Se tiene data como: cambios de trabajo, mudanzas, pérdidas familiares).
19. Smoking (Yes/No) (Fuma): Indica si la personas que formar del dataser fuma (Sí/No).

**Tratamiento de valores outliers**

```python
# Verificamos valores duplicados y los eliminamos
print("Valores duplicados:", df.duplicated().sum())
df.drop_duplicates(inplace=True)

# Seleccionamos solo las columnas numéricas
df_numeric = df.select_dtypes(include=['float64', 'int64'])

# Calculamos el IQR para estas columnas numéricas
Q1 = df_numeric.quantile(0.25)
Q3 = df_numeric.quantile(0.75)
IQR = Q3 - Q1

# Calculamos los límites inferiores y superiores
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtramos las filas con valores atípicos solo en las columnas numéricas
df_filtered_numeric = df_numeric[~((df_numeric < lower_bound) | (df_numeric > upper_bound)).any(axis=1)]

# Ahora, unimos las columnas no numéricas (Gender, Occupation, etc.) al DataFrame filtrado
df_filtered = pd.concat([df_filtered_numeric, df.drop(columns=df_numeric.columns)], axis=1)

# Mostramos el tamaño del dataframe antes y después de filtrar los outliers
print(f"Original DataFrame size: {df.shape}")
print(f"Filtered DataFrame size: {df_filtered.shape}")
```

**Test de normalidad**

```python
from scipy.stats import shapiro
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Test de normalidad para la variable dependiente
stat, p = shapiro(df['Severity of Anxiety Attack (1-10)'])
print(f"Shapiro-Wilk Test: Estadístico={stat}, p-valor={p}")

# QQ-Plot para visualizar la normalidad
sm.qqplot(df['Severity of Anxiety Attack (1-10)'], line='s')
plt.title('QQ-Plot de Severidad de Ansiedad')
plt.show()
```

Al tener p<'0.05', nos indica que la variable Severity of Anxiety Attack (1-10) NO sigue una distribución normal, lo cual imposibilita de cierta manera poder usar la regresión lineal

```python
X = df[['Stress Level (1-10)', 'Sleep Hours', 'Therapy Sessions (per month)', 'Medication']]
X = sm.add_constant(X)
y = df['Severity of Anxiety Attack (1-10)']
model = sm.OLS(y, X).fit()
print(model.summary())
```

El modelo no encuentra evidencia estadística de que la terapia o la medicación reduzcan la severidad de los ataques de ansiedad en esta muestra. La terapia muestra un ligero efecto negativo en la severidad de los ataques , lo que sugiere que podría ayudar, pero el efecto no es significativo por el p = 0.134. El uso de medicación no tiene absolutamente ningún impacto en el modelo peusto que p-valor = 1.000.

```python
# En esta ocasión, se decide crear una nueva variable categórica según horas de sueño
def sleep_category(hours):
    if hours < 5:
        return 'Muy poco'
    elif 5 <= hours < 7:
        return 'Poco'
    elif 7 <= hours <= 9:
        return 'Normal'
    else:
        return 'Mucho'

df_encoded['Sleep category'] = df_encoded['Sleep Hours'].apply(sleep_category)
df_encoded = pd.get_dummies(df_encoded, columns=['Sleep category'], drop_first=True)
```

#**Optimización de modelos**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
# Se define el modelo base
rf = RandomForestClassifier(random_state=42)

# Se tienen los hiperparámetros a probar
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

# Grid Search con validación cruzada de 3 folds
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=3,
    scoring='f1_weighted',
    n_jobs=-1,
    verbose=1
)

# Entrenamos las variables
grid_search.fit(X_train, y_train)
```

**Conclusión sobre la optimización**

```python
# Se desea mostrar los mejores hiperparámetros
print(" Mejores hiperparámetros encontrados:")
print(grid_search.best_params_)

# Se evalua con ello, el modelo optimizado
best_rf = grid_search.best_estimator_
y_pred_opt = best_rf.predict(X_test)

# Se observan las métricas
from sklearn.metrics import classification_report, accuracy_score, f1_score

print("\n Resultados del modelo optimizado:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_opt):.4f}")
print(f"F1 Score: {f1_score(y_test, y_pred_opt, average='weighted'):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_opt))
```
